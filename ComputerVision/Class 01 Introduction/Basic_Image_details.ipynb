```python
import cv2
import matplotlib.pyplot as plt

def image_load_show(img_path):
  # Load an image from a file
  img = cv2.imread(img_path)   # Replace with your own image path
  # OpenCV loads images in BGR format; convert to RGB
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  # Show the image
  plt.imshow(img_rgb)
  plt.title("Original Image")
  plt.axis('off')
  plt.show()
  return img
```

```python
image = image_load_show('bee-8790316_640.jpg')
```

```python
print(f"Image shape: {image.shape}") ## (height, width, 3 RGB channel)
print(f"Data type : {image.dtype}") ## values (0-255)
print(f"Pixel at (100,100): {image[100,100]}")
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("bee-8790316_640.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()
print(f"Image shape: {img.shape}")
```

```python
img[0,0]
```

```python
b,g,r = cv2.split(img)
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(b, cmap= 'gray');plt.title("Blue Channel");plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(g, cmap= 'gray');plt.title("Green Channel");plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(r, cmap= 'gray');plt.title("Red Channel");plt.axis("off")

plt.show()
```

```python
blue_tint = img.copy()
blue_tint[:,:,1] = 0 ## removed green
blue_tint[:,:,2] = 0 ## removed red
plt.imshow(cv2.cvtColor(blue_tint,cv2.COLOR_BGR2RGB))
plt.title("Only Blue CHannel Visible")
plt.axis("off")
plt.show()
```

```python
## Draw shapes on image

canvas = img.copy()
## draw red rectangle
cv2.rectangle(canvas, (50,50),(200,200),(0,0,255), 3)

## draw green circle
cv2.circle(canvas, (300,300), 50, (0,255,0), -1)

## draw text
cv2.putText(canvas, "Hello CV!", (50,400), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0), 2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.title("shapes")
plt.axis("off")
plt.show()
```

```python
## Flip, rortate

flip_h = cv2.flip(img,1)
flip_v = cv2.flip(img,0)
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

crop = img[100:300,100:400]

resized = cv2.resize(img, (200,200))
plt.figure(figsize=(12,6))


plt.subplot(2,3,1);plt.imshow(cv2.cvtColor(flip_h, cv2.COLOR_BGR2RGB));plt.title("Flip H");plt.axis("off")

plt.subplot(2,3,2);plt.imshow(cv2.cvtColor(flip_v, cv2.COLOR_BGR2RGB));plt.title("Flip V");plt.axis("off")

plt.subplot(2,3,3);plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB));plt.title("Rotate 90");plt.axis("off")

plt.subplot(2,3,4);plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB));plt.title("Crop");plt.axis("off")

plt.subplot(2,3,5);plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB));plt.title("Resize");plt.axis("off")
plt.show()
```

```python
## gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_,binary = cv2.threshold(gray, 125,255, cv2.THRESH_BINARY)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1);plt.imshow(gray,cmap='gray');plt.title("GrayScale");plt.axis("off")
plt.subplot(1,2,2);plt.imshow(binary,cmap='gray');plt.title("Binary");plt.axis("off")
plt.show()
```

```python
negative = 255-img
plt.imshow(cv2.cvtColor(negative,cv2.COLOR_BGR2RGB))
plt.title("Negative Image Effect")
plt.axis("off")
plt.show()
```

```python
### This is how color is represented digitally — three separate layers stacked together (RGB).
### Each layer stores how much of that color to show.

r, g, b = img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2]

fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(r, cmap='Reds')
axs[0].set_title('Red Channel')
axs[1].imshow(g, cmap='Greens')
axs[1].set_title('Green Channel')
axs[2].imshow(b, cmap='Blues')
axs[2].set_title('Blue Channel')

for ax in axs:
    ax.axis('off')
plt.show()
```
