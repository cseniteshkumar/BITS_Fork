```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import FastICA
```

```python
np.random.seed(0)
```

```python
n_samples = 2000
time = np.linspace(0, 8 , n_samples)

#Independent sources (true signals)
engine_vibration = np.sin(2 * time)
road_noise = np.sign(np.sin(3*time))
sensor_noise = np.random.normal(size=n_samples)

S = np.c_[engine_vibration, road_noise, sensor_noise]
S /= S.std(axis=0)
```

```python
A = np.array([
    [1,1,1],
    [0.5,2,1],
    [1.5,1,2]

])

X = S @ A.T  #mixed signals
```

```python
ica = FastICA(n_components=3, random_state=42)
S_estimated = ica.fit_transform(X)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3,1,1)

plt.title('Original Independent Signals')
plt.plot(S)

plt.subplot(3,1,2)
plt.title('Mixed Sensor Signals')
plt.plot(X)

plt.subplot(3,1,3)
plt.title('Recovered Signals using ICA')

plt.plot(S_estimated)


plt.tight_layout()
plt.show()
```
