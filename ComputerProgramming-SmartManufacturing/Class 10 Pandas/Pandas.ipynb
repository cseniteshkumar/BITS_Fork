```python
drinks[drinks["continent"]=="Africa"].beer_servings.mean()
```

```python
## calculate mean beer servings for each continent

drinks.groupby("continent").beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','mean','min','max'])
```

```python
drinks.groupby('continent')[["beer_servings","spirit_servings","wine_servings","total_litres_of_pure_alcohol"]].mean()
```

```python
drinks.groupby('continent')[["beer_servings","spirit_servings","wine_servings","total_litres_of_pure_alcohol"]].mean().plot(kind='bar')
```
