```python
'''
  What is OOP? -- Paridgm that organizes software design around objects rather than functions and logic.
  Some concepts:
    * Class - Blueprint for creating objects
    * Object - Instance of class (Like specific machine)
    * Attribute - Data associated with object
    * Method - Functions inside a class that perform operations
'''
```

```python
## Basic Class Example
```

```python
class Machine:
  def __init__(self, name, temperature):
    self.name = name
    self.temperature = temperature

  def check_status(self):
    if self.temperature>90:
      return f"{self.name} is overheating!"
    else:
      return f"{self.name} is operating normally."
```

```python
## creating object - instance of class
m1 = Machine("CNC-001",95)
```

```python
m1.temperature
```

```python
m1.name
```

```python
m1.check_status()
```

```python
## encapsulation - Hides internal object details.
```

```python
class Sensor:
  def __init__(self):
    self.__reading = 0

  def set_reading(self,value):
    if value>=0:
      self.__reading = value

  def get_reading(self):
    return self.__reading
```

```python
s = Sensor()
s.set_reading(88)
print(s.get_reading())
```

```python
## Inheritance - Used to create specialized machine types
```

```python
class Machine:
  def operate(self):
    return "Generic Operation"

class CNC(Machine):
  def operate(self):
    return "Cutting Metal"
```

```python
cnc  = CNC()
print(cnc.operate())
```
