```python
## Use Case 1: Logging Machine data
```

```python
machines = {
    "Machine1": "Running",
    "Machine2": "Idle",
    "Machine3": "Faulty"
}

with open("status_log.txt","w") as file:
  for machine,status in machines.items():
    file.write(f"{machine}: {status}\n")
```

```python
## Use Case 2: Reading Sensor Data from CSV
```

```python
with open("sensor_data.csv","r") as file:
  lines = file.readlines()
  # print(lines)
  for line in lines[1:]:
    timestamp, temp, pressure = line.split("\t")
    print(f"Timestamp: {timestamp}, Temperature: {temp}, Pressure: {pressure}")
  file.close()
```

```python
import time
with open("log.txt","r") as file:
  line = file.readline()
  while line:
    print(f"Read:",line.strip())
    time.sleep(5)
    line=file.readline()
```

```python
'''
Why try-except?: File I/O can fail if file not found, file is locked or in use or data is malformed or permission is denied.
Instead of your program crashing, try-except lets you handle these errors gracefully.
'''
```

```python
a = 4
b = 2
try:
  c = a/b
  print(f"Value of a: {a}, Value of b: {b}, Value of a/b: {c}")
except Exception as e:
  print(f"Exception: {e}")
```

```python
## Best Practices - Handle File not found
```

```python
#read missing file
with open("missing_file.txt","r") as file:
  data = file.read()
  print(data)
  file.close()
```

```python
try:
  with open("missing_file.txt","r") as file:
    data = file.read()
    print(data)
    file.close()
except FileNotFoundError:
  print("File was not found")
```

```python
def read_data(filename):
    with open(filename,"r") as file:
      data = file.read()
      file.close()
    return data
```
